```python
#importing Packages
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
%matplotlib inline
import plotly.express as px
 
print("Sucessfully imported")
```

```python
#LOADING THE DATA
UnempData = pd.read_csv("/content/Unemployment in India.csv") 
print("Sucessfully Loaded the data")
UnempData.head()
```

```python
#To Check Non-Null Values
UnempData.isnull().sum()
```

```python

UnempData.columns= ["Region","Date","Frequency",
               "Estimated Unemployment Rate",
               "Estimated Employed",
               "Estimated Labour Participation Rate",
               "Area"]
plt.style.use('seaborn-whitegrid')
plt.figure(figsize=(10, 10))
sns.heatmap(UnempData.corr())
plt.show()
```

```python
#estimated employed data based on urban and rural area
UnempData.columns= ["Region","Date","Frequency",
               "Estimated Unemployment Rate",
               "Estimated Employed",
               "Estimated Labour Participation Rate",
               "Area"]
plt.title("Unemployment Analysis")
sns.histplot(x="Estimated Employed", hue="Area", data=UnempData)
plt.show()
```

```python
#estimated labour participation rate based on urban and rural areas
UnempData.columns= ["Region","Date","Frequency",
               "Estimated Unemployment Rate",
               "Estimated Employed",
               "Estimated Labour Participation Rate",
               "Area"]
plt.title("Unemployment Analysis")
sns.histplot(x="Estimated Labour Participation Rate", hue="Area", data=UnempData)
plt.show()
```

```python
#Areawise estimated unemployment Rate 
UnempData.columns= ["Region","Date","Frequency",
               "Estimated Unemployment Rate",
               "Estimated Employed",
               "Estimated Labour Participation Rate",
               "Area"]
plt.title("Unemployment Analysis")
sns.histplot(x="Estimated Unemployment Rate", hue="Area", data=UnempData)
plt.show()
```

```python
#Regionwise estimated unemployment Rate
UnempData.columns= ["Region","Date","Frequency",
               "Estimated Unemployment Rate",
               "Estimated Employed",
               "Estimated Labour Participation Rate",
               "Area"]
plt.title("Unemployment Analysis")
sns.histplot(x="Estimated Unemployment Rate", hue="Region", data=UnempData)
plt.show()
```
